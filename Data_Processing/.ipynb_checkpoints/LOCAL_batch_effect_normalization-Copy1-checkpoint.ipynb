```python
def scale_to_min_max(arr, newmin, newmax):
    newarr = []
    for row in arr:
        newarr.append((row - np.min(row))/(np.max(row)-np.min(row)))
    return np.array(newarr)

varz = {}
newX = np.zeros(X.shape)
for fname in np.unique(labels[:,1]):
    Xtemp =  X[np.where(labels[:,1] == fname)[0]]
    newX[np.where(labels[:,1] == fname)[0]] = scale_to_min_max(Xtemp, 0, 1)
    varz[fname] = np.var(Xtemp,axis=0)
#X = newX
```

```python
path_to_data = "/Users/nbbwang/Documents/AD_data/"
HBTRC_GEO_CER_covars_preformated = pd.read_csv(path_to_data + "HBTRC_GEO/GSE44768_series_matrix.txt", delimiter="\t", skiprows=50, low_memory=False)[:52]
HBTRC_GEO_PFC_covars_preformated = pd.read_csv(path_to_data + "HBTRC_GEO/GSE44770_series_matrix.txt", delimiter="\t", skiprows=50, low_memory=False)[:52]
HBTRC_GEO_VC_covars_preformated = pd.read_csv(path_to_data + "HBTRC_GEO/GSE44771_series_matrix.txt", delimiter="\t", skiprows=50, low_memory=False)[:52]

def clean_HBTRC_covars(HBTRC_covars_preformated):

    # get names of samples from second row of unedited data
    samplenames = HBTRC_covars_preformated.values[17,1:]
    geo_accs = HBTRC_covars_preformated.values[0,1:]

    # get relevant covariates rows from data (they all have the format "variable: value")
    covars_unedited = HBTRC_covars_preformated.iloc[19:27].values[:,1:]

    # extract variable names from "variable: value" (variable is the same across the row, so just need to look at first column)
    covars_titles = ["sample"] + ["geo_accession"] + [x.split(": ")[0] for x in covars_unedited[:,0]]

    # exctract values from "variable: value" from each cell. Transpose the data so info is in the form sample by variable
    # then add in sample names and geo accession numbers
    covars_values = np.array([[x.split(": ")[1] for x in arr] for arr in covars_unedited]).T
    covars_values = np.hstack([geo_accs.reshape(-1,1), covars_values])
    covars_values = np.hstack([samplenames.reshape(-1,1),covars_values])

    return pd.DataFrame(covars_values, columns=covars_titles)

HBTRC_GEO_CER_covars = clean_HBTRC_covars(HBTRC_GEO_CER_covars_preformated)
HBTRC_GEO_PFC_covars = clean_HBTRC_covars(HBTRC_GEO_PFC_covars_preformated)
HBTRC_GEO_VC_covars = clean_HBTRC_covars(HBTRC_GEO_VC_covars_preformated)
```
